```python
import duckdb
dataset = duckdb.read_csv('vendas_completo.csv')
```

# **Analise de vendas completa**

- Nesse notebook, realizaremos as análises que necessitam do relatório de vendas completo, com a descrição dos itens vendidos, o códgio de estoque, quantidade e preço individual de cada item.

## **Buscando quantidade de itens vendidos**

- Feitas as devidas limpezas, podemos começar a conferir informações relevantes. A primeira delas, que pode fazer muita diferença no planejamento de estoque e em estratégia de vendas é saber como andam as vendas dos itens disponibilizados.

```python
query = f"""
    SELECT 
        Description, SUM(Quantity) AS TotalQuantity
    FROM 
        dataset
    GROUP BY
        Description 
    ORDER BY
        TotalQuantity DESC
"""
resultado = duckdb.sql(query).df()
display(resultado)
```

## **Item mais vendido (por compra)**

- Agora podemos selecionar o item mais vendido em uma única compra

```python
query = f"""
    SELECT 
        *
    FROM 
        dataset
    ORDER BY
        Quantity DESC
"""
resultado = duckdb.sql(query).df()
display(resultado)
```
