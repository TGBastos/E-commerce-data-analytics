## - **How many rows we are working?**

```python
import duckdb

database = 'complete_ecommerce_data.csv'
query = f"""
    SELECT 
        COUNT(*) 
    FROM 
        read_csv_auto({database}, encoding='latin-1')
"""
resultado = duckdb.sql(query)
display(resultado.df())

```

## - **How many unique orders we have for real?**

```python
query = f"""
    SELECT 
        COUNT(DISTINCT InvoiceNo) 
    FROM 
        read_csv_auto({database}, encoding='latin-1')
"""
resultado = duckdb.sql(query).show()
```

## - **Everything**

```python
query = f"""
    SELECT 
        * 
    FROM 
        read_csv_auto({database}, encoding='latin-1')  
    ORDER BY 
        InvoiceNo ASC"""
resultado = duckdb.sql(query).show()
```

## - **Lets see all information of a same Invoice in a Row?**

```python
query = f"""
    SELECT 
        InvoiceNo,
        InvoiceDate, 
        CustomerId,
        Country, 
        COUNT(Description), 
        SUM(Quantity), 
        ROUND(SUM(UnitPrice)), 
        ROUND(SUM(Quantity)*AVG(UnitPrice)) AS TotalValue 
    FROM 
        read_csv_auto({database}, encoding='latin-1') 
    GROUP BY 
        InvoiceNo, InvoiceDate, CustomerId, Country  
    ORDER BY 
        InvoiceNo ASC"""
resultado = duckdb.sql(query).show()
```
